```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

### 图像阈值

```python
img= cv2.imread('cat.jpg')
img_gray= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img_gray, cmap='gray')
```

```python
thresh, ret1= cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)
thresh, ret2= cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY_INV)
thresh, ret3= cv2.threshold(img_gray, 127, 255, cv2.THRESH_TRUNC)
thresh, ret4= cv2.threshold(img_gray, 127, 255, cv2.THRESH_TOZERO)
thresh, ret5= cv2.threshold(img_gray, 127, 255, cv2.THRESH_TOZERO_INV)

titles= [ "Original", "BINARY", "BINARY_INV", "TRUNC", "TOZERO", "TOZERO_INV" ]
images= [ img_gray, ret1, ret2, ret3, ret4, ret5 ]

fig, axs= plt.subplots(2, 3, figsize=(10, 6))
for i in range(6):
    axs[i//3, i%3].imshow(images[i], cmap='gray')
    axs[i//3, i%3].set_title(titles[i])

```

### 图像平滑处理

```python
img= cv2.imread('lenanoise.png')
```

```python
cv2.imshow("lena", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 均值滤波
img_blur= cv2.blur(img, (3, 3))

cv2.imshow("lena", img_blur)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 方框滤波
# 把3x3的区域内的像素值相加，如果normalize=True，则除以9(和均值滤波的效果是一样的)，如果normalize=False，则除以1
img_box= cv2.boxFilter(img, -1, (3, 3), normalize=True)

cv2.imshow("lena", img_blur)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 高斯滤波
# kernal的数值满足高斯分布，更重视中间的点
img_gaussian= cv2.GaussianBlur(img, (5, 5), 0) # 0 means determiane sigmaX automatically

cv2.imshow("lena", img_gaussian)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 中值滤波
# 相当于用中值代替
img_median= cv2.medianBlur(img, 5)

cv2.imshow("lena", img_median)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 将所有图在水平方向拼接
res= np.hstack((img, img_blur, img_gaussian, img_median))

cv2.imshow("lena", res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
