### Canny边缘检测
- 使用高斯滤波器，以平滑图像，滤除噪声
- 计算图像中每个像素点的梯度强度和方向 <br>
　角度= arc tan(Gy/Gx)
- 应用非极大值抑制，以消除边缘检测带来的杂散响应 <br>
　比较检测到的边缘点，和离它最近的两个点的梯度强度，然后只取最大的为边缘点
- 应用双阈值检测来确定真实的和潜在的边缘<br>
1) 梯度值>maxVal则处理为边界
1) minVal<梯度值<maxVal，连有边界则保留，否则舍弃
1) 梯度值<minVal, 则舍弃
即当上下阈值都比较小时，可以检测到更多细节
- 通过抑制孤立的弱边缘最终完成边缘检测

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img= cv2.imread('lenanoise.png', cv2.IMREAD_GRAYSCALE)
```

```python
img1= cv2.Canny(img, 120, 250)
img2= cv2.Canny(img, 80, 120)
img12= np.hstack((img1, img2))
```

```python
plt.imshow(img12, cmap='gray')
```
